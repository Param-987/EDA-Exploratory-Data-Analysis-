# What is RFM Analysis?
The RFM model is a customer segmentation technique. RFM stands for Recency, Frequency, and Monetary value, each corresponding to some key customer trait. These RFM metrics are important indicators of a customer's behavior because frequency and monetary value affects a customer's lifetime value, and recency affects retention, a measure of engagement.

Monetary : This factor reflects how much a customer has spent with the brand during a particular period of time.        
Recency  : How much time has elapsed since a customerâ€™s last activity or transaction with the brand.                    
Frequency: How often has a customer transacted or interacted with the brand during a particular period of time

```python
# """Monetary is provided by coloumn name REVENUE
# Frequency is provided by Total_ORDERS
# Recency is provided by DAYSSINCELASTORDER( but lets check is it so)"""
```

```python
# Getting the max date in last_order_date , it will show the most recent order made.
todays_date = df['LATEST_ORDER_DATE'].max()
```

```python
#calculating the recency by subtracting the last order date of each customer with the todays_date 
rfm = pd.DataFrame() # we will perform anaylsis on this df
rfm['recency'] = df['LATEST_ORDER_DATE'].apply(lambda x: (todays_date - x).days+1)
np.sum(rfm['recency'] == df['DAYSSINCELASTORDER'])
# if the calculated recency is same as DAYSSINCELASTORDER then sum will be 5k
```

KMeans clustering is an unsupervised machine learning technique where we cluster/segment the data points into groups based on their similarity among them. We can find the optimal numbers of cluster to divide our customers into , to perform analysis.

```python

k_rfm = pd.DataFrame()
k_rfm[['recency','frequency','monetary']] = df[['DAYSSINCELASTORDER','TOTAL_ORDERS','REVENUE']]
k_rfm.head()
```

```python
"""Since the KMeans cluster is a distance based algorithm , we should transform our data into single scale to get the 
optimal results from KMeans Algorithm"""
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
k_rfm = scaler.fit_transform(k_rfm)

# We will run the KMeans algo over 10 different cluster and get the best one from it using Elbow curve
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(k_rfm)
    sse.append(km.inertia_) # km.intertia_ give the errors(sum of sqr of dist between points and centroids)
```

```python
sse # the optimal errors on cluster(1-10)
```

```python
""" The error will keep on decreasing and will become 0 when no of cluster == data points , best number of cluster is
when we found the elbow"""
plt.xlabel('n_cluster')
plt.ylabel('SSE')
plt.title('Elbow Curve KMeans')
plt.plot(k_rng,sse,marker='o')
```

The above plot shows we should divide customers in 3 clusters , but 3 cluster won't let us divide customers into segment effectively so we will use 4 cluster.

```python
# Creating a RFM dataframe for RFM analysis
rfm[['CustomerID','recency','frequency','monetary']]=df[['CustomerID','DAYSSINCELASTORDER','TOTAL_ORDERS','REVENUE']]
```

```python
# To draw the barplot on to visualize the results
def DrawBarPlot(x,y,xl,yl,title):
    def add_value_label(x_list,y_list):
        for i in range(1, len(x_list)+1):
            plt.text(i-1,y_list[i-1]/2,f"{y_list[i-1]/50}%({y_list[i-1]})",ha="center"
                     ,fontdict={'color':'#000000','fontsize':10,'fontfamily':'monospace'})
            
    plt.figure(figsize=(15,6))
    x , y = list(x) ,list(y)
    sns.barplot(x = x , y = y ,color="#6A67CE")
    plt.xlabel(xl,fontdict={'fontsize':16,'fontfamily':'monospace'})
    plt.ylabel(yl,fontdict={'fontsize':16,'fontfamily':'monospace'})
    plt.title(title,fontdict={'fontsize':16,'fontfamily':'monospace'})
    add_value_label(x,y)
    plt.tick_params(axis="x", labelsize=10, labelrotation=40, labelcolor="#16003B")
    plt.tick_params(axis="y", labelsize=10,labelcolor="#000000")

def DrawColumnsBarplot(col,xl,yl,title):
    a = rfm.groupby(col).size()
    a.to_dict()
    DrawBarPlot(a.keys(),a.values,xl,yl,title)
```

```python
# Diving the whole dataset into 4 quartiles(Kmeans help us to decide) 
quartiles = rfm.quantile(q=[0.25,0.50,0.75])
print(quartiles, type(quartiles))
```

```python
quartiles=quartiles.to_dict()
quartiles
```

We will segment our customers based on above scale                                   
Lower the recency, Higher the Frequency and Monetary , important the  customer is

```python
#For recency class where ranking is alloted based on above scaling(lesser the value , lower the rank)
def RClass(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
    
## for Frequency and Monetary class based on above scaling(more the value , lower the rank) 

def FMClass(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1    
    
```

```python
rfm.head()
```

```python
# transform RFM values to get corresponding RFM scores
rfm['R_score'] = rfm['recency'].apply(RClass, args=('recency',quartiles,))
rfm['F_score'] = rfm['frequency'].apply(FMClass, args=('frequency',quartiles,))
rfm['M_score'] = rfm['monetary'].apply(FMClass, args=('monetary',quartiles,))
```

```python
#RFM_class is the string concatenation of RFM score
rfm['RFM_Class'] = rfm.R_score.astype(str) + rfm.F_score.astype(str) + rfm.M_score.astype(str)

#RFM_Score is the integer addition of RFM score(range between 3-12)
rfm['RFM_Score'] = rfm.R_score + rfm.F_score + rfm.M_score
```

```python
rfm['RFM_Class'].unique() # All the different values of RFM_Class
```

```python
# Draw barplot on RFM Score vs frequcny of customers having that score
DrawColumnsBarplot('RFM_Score','Customers RFM_Score','Frequency of Customers','Customer RFM Analysis')
```

```python
# MApping the customers based on their RFM_Class to a human understanble scale
seg_map = {
    r'11[1-2]':'Champion', 
    r'14[1-2]':'High Spending New Customer',
    r'14[3-4]':'New_Customer',
    r'[1-2][1-2][1-2]':'Loyal_Customer&Spend_above_avg', 
    r'[1-2][1-4][1-4]':'Potential and Promising', 
    r'[3-4][1-2][1-2]':'Cant_loose',
    r'3[1-4][1-4]': 'Need_attention',
    r'4[1-2][1-4]':'At_risk',
    r'4[3-4][1-4]': 'Hibernating',
}
# Above conditions are regular expression matching
# r11[1-2] mean customers having 111 and 112 RFM_CLass
```

```python
# Each Scale meaning
""" 
Champions :                  They transacted recently, do so often and spend more than other customers.
High Spending New Customer : As the name suggest , new Customer who spends alot.
Loyal customers :            Buy on a regular basis. Responsive to promotions.
New customers :              Bought most recently, but not often.
Potential and Promising :    Recent shoppers, but havenâ€™t spent much.
Can't loose:                 The ones who spend alot and buy frequently but last transaction was long back.
Need Attention:              Below average recency and frequency. Will lose them if not reactivated.
At risk:                     Some time since theyâ€™ve purchased. Need to bring them back!
Hibernating:                 Last purchase was long back and low number of orders. May be lost.
"""
```

```python
# Segmenting customers on above scale 
rfm['segment'] = rfm['RFM_Class'].replace(seg_map, regex=True)
```

```python
# droping the columns not required
df.drop(['TOTAL_ORDERS','REVENUE','FIRST_ORDER_DATE','AVERAGE_ORDER_VALUE','LATEST_ORDER_DATE'],axis=1,inplace=True)
```

```python
# Seperate dataframes for each section
df_weekname = pd.DataFrame()
df_weekdays = pd.DataFrame()
df_time = pd.DataFrame()
```

```python

def Seperate_df():
    weekname = ['MONDAY','TUESDAY','WEDNESDAY','THURSDAY','FRIDAY','SATURDAY','SUNDAY']
    
    weekname_order = list(map(lambda x : x+'_ORDERS' ,weekname))
    weekname_revenue = list(map(lambda x : x+'_REVENUE' ,weekname))
    
    df_weekname[weekname_order] = df[weekname_order]
    df_weekname[weekname_revenue] = df[weekname_revenue]
    for name in df.columns:
        if name.startswith('WEEK'):
            df_weekdays[name] = df[name]
        if name.startswith('TIME'):
            df_time[name] = df[name]
    df.drop(df_weekname.columns,axis=1,inplace=True)
    df.drop(df_weekdays.columns,axis=1,inplace=True)
    df.drop(df_time.columns,axis=1,inplace=True)
    df_weekdays['CustomerID'] = df['CustomerID']
    df_weekname['CustomerID'] = df['CustomerID']
    df_time['CustomerID'] = df['CustomerID']
    
#     print(df.columns,df_weekdays.columns,df_weekname.columns,df_time.columns)
    
# weekname.extend(['CustomerID'])
# df_weekname = df[weekname]
```

```python
df_anal_weekdays = df_weekdays.merge(right=rfm[['CustomerID','segment']],how='inner',on='CustomerID')
df_anal_weekname = df_weekname.merge(right=rfm[['CustomerID','segment']],how='inner',on='CustomerID')
df_anal_time = df_time.merge(right=rfm[['CustomerID','segment']],how='inner',on='CustomerID')
```

```python
# this fucntion will create the barplot for each segment of customer (champion, hibernating ,....)
def Analyse(seg):
    df_weekdays_seg = df_anal_weekdays[df_anal_weekdays['segment']==seg].mean(numeric_only=True)
    df_weekname_seg = df_anal_weekname[df_anal_weekname['segment']==seg].mean(numeric_only=True)
    df_time_seg = df_anal_time[df_anal_time['segment']==seg].mean(numeric_only=True)

    plt.figure(figsize=(20,20))

    plt.subplot(3,2,1)
    df_weekdays_seg[:4].plot.bar()
    plt.xticks(rotation=20)

    plt.subplot(3,2,2)
    df_weekdays_seg[4:-1].plot.bar()
    plt.xticks(rotation=20)

    plt.subplot(3,2,3)
    df_weekname_seg[:7].plot.bar()
    plt.xticks(rotation=20)

    plt.subplot(3,2,4)
    df_weekname_seg[7:-1].plot.bar()
    plt.xticks(rotation=20)

    plt.subplot(3,2,5)
    df_time_seg[:4].plot.bar()
    plt.xticks(rotation=20)

    plt.subplot(3,2,6)
    df_time_seg[4:-1].plot.bar()
    plt.xticks(rotation=20)

    # plt.subplot_tool()
    plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=.3, hspace=.3)
    plt.show()

Analyse('Hibernating')
```

Using above fucntion we can see how the each section of customer perform on the basis of weekdays, weekname,time. It will also show when is the customer more liabel to be active, thus company should prefer this result in their marker strategy.
